```python
# Importando Bibliotecas de Análise de Dados:
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando arquivo csv do desafio usando método do pandas:
df = pd.read_csv('/content/dados_rh.csv')
df
```

```python
# Vendo a descrição e as informações do Dataframe:
df.describe()
```

```python
# Verificando quantas linhas e colunas temos no DataFrame
n_linhas, n_colunas = df.shape
print(f'O DataFrame tem {n_linhas} linhas e {n_colunas} colunas.')
```

```python
# Analisando se há valores nulos no DataFrame
colunas_nulos = df.columns[df.isna().any()]

if colunas_nulos is True:
  print("Sim, há dados nulos no banco")
  quantidade_nulos_coluna = df[colunas_nulos].isna().sum()

  for coluna, qtd in quantidade_nulos_coluna.items():
      print(f"A coluna '{coluna}' tem {qtd} dados nulos.")
else:
 print("Não há dados nulos no banco.")
```

```python
# Compreendendo quantos tipos de dados temos no DataFrame
tipos_dados = df.dtypes

qtd_tipos_de_dados = tipos_dados.value_counts().reset_index()
qtd_tipos_de_dados.columns = ['Tipo de Dado', 'Quantidade']

print("Quantidade de cada tipos de dados no DataFrame:")
display(qtd_tipos_de_dados)
```

```python
# Agrupando dados pela quantidade de funcionários em cada setor
qtd_departamento = df['Department'].value_counts()

# Plotando os gráficos
plt.figure(figsize=(8, 4))
qtd_departamento.plot(kind='bar')
plt.title('Quantidade de Funcionários por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Quantidade de Funcionários')
plt.xticks(rotation=45)

plt.show()
```

```python
# Agrupando por a quantidade de funcionários para cada nível de escolaridade
qtd_educacao = df['Education'].value_counts().sort_index()

# Plotando os gráficos
plt.figure(figsize=(8, 4))
qtd_educacao.plot(kind='line', marker='o', linestyle='-', linewidth=2)
plt.title('Quantidade de Funcionários por Nível de Escolaridade')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Quantidade de Funcionários')
plt.grid(True)

plt.show()
```

```python
# Agrupando dados pela quantidade de funcionários que tem ou não algum atrito na empresa
agrupando_attrition = df['Attrition'].value_counts()

# Plotando os gráficos
plt.figure(figsize=(6, 6))
plt.pie(agrupando_attrition, labels=agrupando_attrition.index, autopct='%1.1f%%')
plt.title('Funcionários com ou sem Atrito na Empresa')

plt.show()
```
